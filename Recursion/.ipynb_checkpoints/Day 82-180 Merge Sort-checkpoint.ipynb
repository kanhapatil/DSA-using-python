### 1. Merge Sort



```python
def merge(arr, start, mid, end):
    temp = [None] * (end - start + 1)
    left, right, index = start, mid + 1, 0

    print(f"Merge: start={start}, mid={mid}, end={end}")

    while left <= mid and right <= end:
        print(f"  Comparing arr[{left}]={arr[left]} and arr[{right}]={arr[right]}")

        if arr[left] <= arr[right]:
            temp[index] = arr[left]
            index += 1
            left += 1
        else:
            temp[index] = arr[right]
            index += 1
            right += 1

    while left <= mid:
        temp[index] = arr[left]
        index += 1
        left += 1

    while right <= end:
        temp[index] = arr[right]
        index += 1
        right += 1

    index = 0
    for i in range(start, end + 1):
        arr[i] = temp[index]
        index += 1

    print("  Merged array:", arr[start:end + 1])

    return arr


def mergeSort(arr, start, end):
    # Base case
    if start == end:
        return

    mid = start + (end - start) // 2

    print("Divided arrays:", arr[start:mid + 1], arr[mid + 1:end + 1])

    mergeSort(arr, start, mid)
    mergeSort(arr, mid + 1, end)

    merge(arr, start, mid, end)


arr = [85, 24, 63, 45, 17, 31, 96, 50]
start, end = 0, len(arr) - 1
mergeSort(arr, start, end)
```

### 2. Less code for mergeSort

```python
def merge(arr, start, mid, end):
    # Create an empty temporary list to store merged values
    temp = []

    # Initialize pointers for the left and right subarrays
    left, right = start, mid + 1

    # Merge elements from the left and right subarrays into temp
    while left <= mid and right <= end:
        if arr[left] <= arr[right]:
            # Append the smaller element from the left subarray
            temp.append(arr[left])
            left += 1
        else:
            # Append the smaller element from the right subarray
            temp.append(arr[right])
            right += 1

    # Append remaining elements from the left subarray (if any)
    temp += arr[left:mid + 1]

    # Append remaining elements from the right subarray (if any)
    temp += arr[right:end + 1]

    # Update the original array with the sorted values from temp
    arr[start:end + 1] = temp

    # Return the sorted array
    return arr


def mergeSort(arr, start, end):
    # Base case: If the subarray has only one element, it's already sorted
    if(start < end):

        # Calculate the middle index of the current subarray
        mid = start + (end - start) // 2

        # Recursively sort the left and right subarrays
        mergeSort(arr, start, mid)
        mergeSort(arr, mid + 1, end)

        # Merge the sorted left and right subarrays
        return merge(arr, start, mid, end)


# Example usage:
arr = [85, 24, 63, 45, 17, 31, 96, 50]
start, end = 0, len(arr) - 1

# Perform merge sort on the entire array
result = mergeSort(arr, start, end)

print("Merge sort:", result)
```

### 3. Count Inversions

```python
def merge(arr, start, mid, end):
    temp = []
    invCount = 0

    left, right = start, mid + 1

    while left <= mid and right <= end:
        if arr[left] <= arr[right]:
            temp.append(arr[left])
            left += 1
        else:
            temp.append(arr[right])
            right += 1
            invCount += (mid - left + 1)

    temp += arr[left:mid + 1]
    temp += arr[right:end + 1]

    arr[start:end + 1] = temp

    return invCount


def mergeSort(arr, start, end):
    invCount = 0

    if start < end:
        mid = start + (end - start) // 2

        invCount += mergeSort(arr, start, mid)
        invCount += mergeSort(arr, mid + 1, end)
        
        invCount += merge(arr, start, mid, end)
        
    return invCount

# Example usage:
arr = [2, 4, 1, 3, 5]
start, end = 0, len(arr) - 1

result = mergeSort(arr, start, end)

print("Merge sort:", result)
```
